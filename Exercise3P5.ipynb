P5> Plot a graph showing the relationship between PPN and number of Bedrooms.

-- Initial steps

```python
import pandas as pd
df = pd.read_excel("cleaned_data.xlsx")
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.head()
```

-- Plotting the relationship Graph usng Scatter plot

```python
ax=df.plot(kind='scatter', x='Beds', y='PPN', color='blue', alpha=0.7, edgecolors='black', figsize=(8, 5))
ax.set_xlabel("Number of Bedrooms")
ax.set_ylabel("Price Per Night (PPN)")
ax.set_title("Relationship between PPN and Number of Bedrooms")
ax
```

# Reasoning:
- A scatter plot is chosen because it effectively visualizes the relationship between two numerical variables.
- It helps to identify trends, correlations, and outliers in the data, showing how PPN changes with the number of bedrooms.



# Inference and Observations
- From the scatter plot, we can observe that as the number of bedrooms increases, the PPN generally tends to rise.
- However, there are some variations due to factors like location and property type, which may also influence pricing.
