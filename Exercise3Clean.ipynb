
## Room Data Cleaning


- Load "cap_rooms.csv" file using Pandas module


```python
import pandas as pd
folder="data"
filename="cap_rooms.csv"
path=f'{folder}/{filename}'
df=pd.read_csv(path)
```

-- Importing matplotlib module to plot and observe chart behaviours

```python
import matplotlib.pyplot as plt
```

```python
fig,ax=plt.subplots(2,2)
cols=['Nights', 'Beds', 'Baths', 'Reviews']
df.Nights.plot.kde(ax=ax[0,0], color='blue')
df.Beds.plot.kde(ax=ax[0,1], color='green')
df.Baths.plot.kde(ax=ax[1,0], color='red')
df.Reviews.plot.kde(ax=ax[1,1], color='purple')

# adding titles to each chart for clarity

ax[0, 0].set_title('Density Plot of Nights')
ax[0, 0].set_xlabel('Nights')
ax[0, 0].set_ylabel('Density')
handles, labels = ax[0, 0].get_legend_handles_labels()

ax[0, 1].set_title('Density Plot of Beds')
ax[0, 1].set_xlabel('Beds')
ax[0, 1].set_ylabel('Density')
handles, labels = ax[0, 1].get_legend_handles_labels()

ax[1, 0].set_title('Density Plot of Baths')
ax[1, 0].set_xlabel('Baths')
ax[1, 0].set_ylabel('Density')
handles, labels = ax[1, 0].get_legend_handles_labels()

ax[1, 1].set_title('Density Plot of Reviews')
ax[1, 1].set_xlabel('Reviews')
ax[1, 1].set_ylabel('Density')
handles, labels = ax[1, 1].get_legend_handles_labels()

# fixing the charts' layout
plt.tight_layout(rect=[0, 0, 1, 0.96])


# Title of entire chart
fig.suptitle('KDE Plots for Nights, Beds, Baths & Reviews', fontsize=14)
```

- Filling missing data in Nights and Reviews column with average value since Nights showed standard deviation in chart
- Filling missing data in Beds and Baths column with median value since Beds showed skewed deviation in chart

```python
df['Nights']=df.Nights.fillna(df.Nights.mean())
df['Beds']=df.Beds.fillna(df.Beds.median())
df['Baths']=df.Baths.fillna(df.Baths.median())
df['Reviews']=df.Reviews.fillna(df.Reviews.mean())
df.head()
```

-- Checking if any null values are present in any numeric column

```python
df.isna().sum()
```

-- 2 Null values are present only in Location column , which is neither numeric and nor categorical so we can fill it wit any value like 'Unknown' since dropping these records would drop data from other columns which might affect analysis and calculation for essential columns.

```python
df[df['Location '].isnull()]
```

```python
df['Location ']=df['Location '].fillna('Unknown')
```

```python
df.info()
```

-- Recent column has improper datatype, which needs to be fixed

```python
df.Recent
```

-- importing dateutil module's parser funtion to efficiently handle date anomalies

```python
from dateutil import parser
```

```python
def fix_date(date_str):
        return pd.to_datetime(parser.parse(date_str)).strftime('%Y-%m-%d')
```

```python
df['Recent']=df['Recent'].apply(fix_date)
df.head()
```

-- fix Datatype for Recent

```python
df['Recent']=pd.to_datetime(df['Recent'])
```

```python
df.info()
```

-- checking City Column anomalies

```python
df.City.unique()
```

-Updating City Column with appropriate names

```python
df['City']=df.City.replace({
        'Bangalor$': 'Bangalore',
        'Delhi ':'Delhi',
        'Hydrabad':'Hyderabad'
    },
    regex=True
)
```

```python
df.City.unique()
```

-- Checking PPN Column for anomalies

```python
df.PPN.unique()
```

-- Fixing PPN anomalies and updating its datatype to integer

```python
df['PPN']=df.PPN.replace(
    {
        '\u20b9': '', # u20b9 in code for rupee symbol
        ',':'',
        'One Thousand Five Hundred':1500,
        'Two Thousand':2000,
        'Eight Hundred':800
    },
    regex=True
).astype(int)
```

```python
df.PPN.unique()
```
