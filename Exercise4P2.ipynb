-- initial operations import necessary modules and read data from the three files

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df_bollywood = pd.read_excel("bollywood_movies.xlsx")
df_tollywood = pd.read_excel("tollywood_movies.xlsx")
df_indian = pd.read_excel("indian_movies.xlsx")
```

```python
df_bollywood,df_indian,df_tollywood
```

-- Combine the three datasets

```python
df_combined = pd.concat([df_bollywood, df_tollywood, df_indian])
df_combined.head()
```

-- extracting ecessary information from combined table for plotting

-- I am dropping duplicates since combining the three tables had duplicate entries from each table because of same data present in parent table(bollywood) and other child tables

```python
df_pvt = df_combined[['MovieID', 'Title', 'BoxOffice (Crores)']].drop_duplicates()
df_pvt
```

-- Plot the result

```python
ax=df_pvt[['BoxOffice (Crores)']].plot(kind='box', legend=True,title='BoxOffice Distribution')
ax.set_ylabel("BoxOffice in Crores")
```

-- Above boxplot shows outliers present that are beyond the whiskers

--to find the outliers we calculate value between 25% and 75% of data

```python
Q1 = df_pvt['BoxOffice (Crores)'].quantile(0.25)  
Q3 = df_pvt['BoxOffice (Crores)'].quantile(0.75) 
IQR=Q3-Q1
IQR # inter quantile range (formula to find outliers) which reprresnts 50% of data
```

now we need to find the lower and upper bounds for outliers

```python
lower_bound = Q1 - 1.5 * IQR 
upper_bound = Q3 + 1.5 * IQR 
```

-- filtering outliers above upper bound

```python
outliers = df_pvt[df_pvt['BoxOffice (Crores)'] > upper_bound]
outliers
```

-- filtering outliers below lower bound

```python
outliers = df_pvt[df_pvt['BoxOffice (Crores)'] < lower_bound]
outliers
```

-- none present
