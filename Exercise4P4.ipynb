-- initial steps

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_bollywood = pd.read_excel("bollywood_movies.xlsx")
df_tollywood = pd.read_excel("tollywood_movies.xlsx")
df_indian = pd.read_excel("indian_movies.xlsx")
```

```python
df_bollywood.head()
```

-- Analysis 1: Top Performing Director by BoxOffice collection

```python
top5_director=df_bollywood.groupby('Director')['BoxOffice (Crores)'].sum().nlargest(5)
top5_director
```

-- Plot the result

```python
ax=top5_director.plot(kind='bar',legend=True,title='Top Performing Directors by BoxOffice collection')
ax.set_xlabel('')
ax.set_ylabel('')
```

## INFERENCE

S. S. Rajamouli leads by a huge margin – His movies have earned nearly ₹5000 crores, making him the most successful director in terms of box office collections.

Nitesh Tiwari and Rajkumar Hirani follow – With ₹2140 crores and ₹1252 crores respectively, their films have also done exceptionally well, but they are far behind Rajamouli.

Ali Abbas Zafar and S. Shankar complete the top 5 – Their movies have earned around ₹800-₹900 crores, showing that they are successful but not at the same scale as the top three.

Rajamouli’s dominance is clear – His collections are more than double of the second-place director, proving his films attract the largest audiences.

Analysis 2: Actors/Actress who are responsible for higher BoxOffice Collection

```python
top5_actor=df_bollywood.groupby('LeadActor')['BoxOffice (Crores)'].sum().nlargest(5)
top5_actor
```

```python
top5_actress=df_bollywood.groupby('LeadActress')['BoxOffice (Crores)'].sum().nlargest(5)
top5_actress
```

plotting the result in subplots for easy comparison

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))  # Create 1 row, 2 columns of subplots

# actress
top5_actress.plot(kind='bar', ax=ax[0], color='lightcoral', legend=False)
ax[0].set_title("Top 5 Actresses by Box Office Collection")
ax[0].set_xlabel("Lead Actress")
ax[0].set_ylabel("Box Office (Crores)")
ax[0].tick_params(axis='x', rotation=45)

# actors
top5_actor.plot(kind='bar', ax=ax[1], color='royalblue', legend=False)
ax[1].set_title("Top 5 Actors by Box Office Collection")
ax[1].set_xlabel("Lead Actor")
ax[1].set_ylabel("Box Office (Crores)")
ax[1].tick_params(axis='x', rotation=45)

plt.tight_layout() 
plt.show()
```

## INFERENCE

Male actors have earned more at the Box Office compared to female actresses. The highest-grossing male actor has brought in significantly more revenue than the highest-grossing female actress.

Among female leads, Anushka Shetty has the highest total Box Office collection. However, even her total earnings are lower than the top male actor's earnings.

Bollywood’s top-earning actors include names like S. S. Rajamouli and Nitesh Tiwari, who have delivered massive hits. On the other hand, actresses like Anushka Shetty and Fatima Sana Shaikh have also contributed to major blockbusters but with lower total earnings.

This suggests that big-budget and high-earning movies are more often led by male actors. While female-led films perform well, they generally make less in total collections compared to male-led films.
