P1> Calculate ROI ( Return on Investment ) for each movie
	ROI is calculated as : Profit / Budget
	Profit is calculated as : BoxOffice Collections - Budget
	Which are the top 3 movies with highest ROI across all industries?
	Which are the top 3 Hindi movies with highest ROI?
	Which are the top 3 Telugu movies with highest ROI?
	Plot all this information in separate graphs


-- Importing necessary modules

```python
import pandas as pd
import matplotlib.pyplot as plt
```

-- Loading cleaned files in different dataframes for analysis

```python

df_bollywood = pd.read_excel("bollywood_movies.xlsx")
df_tollywood = pd.read_excel("tollywood_movies.xlsx")
df_indian = pd.read_excel("indian_movies.xlsx")
```

```python
df_bollywood.head()
```

```python
df_tollywood.head()
```

```python
df_indian.head()
```

-- To calculate ROI for each dataset/df

-- first calculate profit i.e  boxOffice collection- Budget

```python
df_bollywood['Profit']=df_bollywood['BoxOffice (Crores)']-df_bollywood['Budget (Crores)']
df_bollywood.head()
```

```python
df_tollywood['Profit']=df_tollywood['BoxOffice (Crores)']-df_tollywood['Budget (Crores)']
df_tollywood.head()
```

```python
df_indian['Profit']=df_indian['BoxOffice (Crores)']-df_indian['Budget (Crores)']
df_indian.head()
```

--- Now calculating ROI i.e profit/budget for each df

```python
df_bollywood['ROI']=df_bollywood['Profit']/df_bollywood['Budget (Crores)']
df_bollywood.head()
```

```python
df_tollywood['ROI']=df_tollywood['Profit']/df_tollywood['Budget (Crores)']
df_tollywood.head()
```

```python
df_indian['ROI']=df_indian['Profit']/df_indian['Budget (Crores)']
df_indian.head()
```

-- top3 movies across all industries based on ROI

-- combine all datasets and sort the values on ROI in descending order

-- but before merging, I need to remove index(auto created by excel) to remove redundancy while merging 

```python
df_bollywood = df_bollywood.drop(columns=['index'])
df_indian= df_indian.drop(columns=['index'])
df_tollywood = df_tollywood.drop(columns=['index'])
```

```python
df_bollywood = df_bollywood.reset_index(drop=True)
df_indian= df_indian.reset_index(drop=True)
df_tollywood = df_tollywood.reset_index(drop=True)
```

```python

df_combined = pd.concat([df_bollywood, df_tollywood, df_indian],ignore_index=True).sort_values(by='ROI', ascending=False)
top_3=df_combined.head(3)
top_3
```

-- Now Plot the above result


```python
df_res=top_3[['Title','ROI']]
df_res
```

```python
ax=df_res.plot(
    x='Title', 
    y='ROI', 
    kind='bar', 
    legend=True, 
    title='Top 3 Movies Based on ROI',
    color=['red']
)
ax.set_ylabel("ROI in Percentage")
ax.set_xlabel("Movie Title")
plt.xticks(rotation=30, ha='right') 

```

-- for top3 hindi movies , we need to filter df_combined on the basis of language then sort by ROI in descending order

```python
top_3=df_combined[df_combined['Language']=='Hindi'].sort_values(by='ROI', ascending=False).head(3)
top_3[['Title','ROI']]
```

-- Plot this result

```python
ax=top_3.plot(
    x='Title', 
    y='ROI', 
    kind='bar', 
    legend=True, 
    title='Top 3 Hindi Movies Based on ROI',
    color=['yellow']
)
ax.set_ylabel("ROI in Percentage")
ax.set_xlabel("Movie Title")
plt.xticks(rotation=30, ha='right') 
```

-- top 3 telugu Movies with Highest ROI

```python
top_3=df_combined[df_combined['Language']=='Telugu'].sort_values(by='ROI', ascending=False).head(3)
top_3[['Title','ROI']]
```

-- Plot the result

```python
ax=top_3.plot(
    x='Title', 
    y='ROI', 
    kind='bar', 
    legend=True, 
    title='Top 3 Telugu Movies Based on ROI',
    color=['blue']
)
ax.set_ylabel("ROI in Percentage")
ax.set_xlabel("Movie Title")
plt.xticks(rotation=30, ha='right') 
```
