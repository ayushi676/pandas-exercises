```python
import re
```

```python
df.Date
```

```python
def fix_dates(date_str):
    if re.match(r'\d{2}/\d{2}/\d{4}', date_str):
        date_str = date_str.replace('/', '-')
        date_str = pd.to_datetime(date_str, format='%d-%m-%Y').strftime('%Y-%m-%d')
    elif re.match(r'^\d{1,2}/\d{1,2}/\d{2,4}$', date_str):
        date_str = pd.to_datetime(date_str, format='%d/%m/%y').strftime('%Y-%m-%d')

    return date_str
```

```python
df['Date'].apply(fix_dates)
```

```python
df['Date']=df['Date'].apply(fix_dates)
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
df['Month']=df.Date.dt.month
df.head()
```
